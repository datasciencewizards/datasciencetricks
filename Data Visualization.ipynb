# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Read data

```python
df = pd.read_csv('data/cancer_data.csv')
df.head()
```

# Visualization Begins!!

### Some basic plots first

#### Lets see what is the most common age of cancer diagnosis using bar plot

```python
#To set the figure size of plot
plt.figure(figsize=(25,10))

df['Age at diagnosis'].value_counts().plot.bar(color='red')
```

#### Illustration of line chart using matplotlib

```python
plt.figure(figsize=(25,10))
plt.plot(df["Age at diagnosis"])
plt.show()
```

#### Finding the gender most prone to cancer with pie chart

```python
plt.figure(figsize=(10,7))
plt.pie(df["Gender"].value_counts(),autopct='%1.1f%%',labels = ["male","female"])
plt.show()
```

### Some Complex and interactive plots now!!

#### Finding correlation between different health indicators using clustermap

Role of clustermap is to show the correlation of all the columns present in the dataframe using heat map methodology to determine the inter dependency of the columns in the dataframe

```python
sns.clustermap(df.iloc[:,:7].corr(),figsize=(15,10))
```

### Finding out if younger people can have cancer with boxplot

Box plot is used to check the spread of numerical values within a column and has intensive application in determining the outliers in a column

```python
sns.boxplot(df["Age at diagnosis"])
```

### Violinplot to find most common gender and age of cancer patients 

Violin plot is a method of plotting numeric data. It is similar to a box plot, with the addition of a rotated kernel density plot on each side. Violin plots are similar to box plots, except that they also show the probability density of the data at different values, usually smoothed by a kernel density estimator.

```python
plt.figure(figsize=(11,6))
sns.set_style("whitegrid") 
  
sns.violinplot(x = 'Gender', y = 'Age at diagnosis', data = df) 
```

### Finding outliers with swarmplot

Swarmplot is a good complement to a box or violin plot in cases where you want to show all observations along with some representation of the underlying distribution. This plot can be used inter dependency between 2 or more variables(be it numerical or categorical)

```python
sns.swarmplot(x="Ascites degree", y="Age at diagnosis",hue="Alcohol",
              palette=["r", "c", "y"], data=df)
```

### Jointplot to show "smoking causes cancer"

Seaborn's jointplot displays a relationship between 2 variables (bivariate) as well as 1D profiles (univariate) in the margins. This plot is a convenience class that wraps JointGrid and covers functionality of two plots and thus helps in analysing in more dimensions

```python
sns.jointplot(df["Packs of cigarets per year:"],df["Age at diagnosis"], kind="kde", height=7, space=0)
```
